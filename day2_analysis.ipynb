```python
import pandas as pd

df = pd.read_csv("student.csv")

# 1. Average marks
avg = df["Marks"].mean()

# 2. Topper
topper = df.loc[df["Marks"].idxmax()]

# 3. Subject-wise average
subject_avg = df.groupby("Subject")["Marks"].mean()

# 4. Students above 85
above_85 = df[df["Marks"] > 85]

print("Average Marks:", avg)
print("Topper:", topper["Name"])
print("\nSubject-wise Average:\n", subject_avg)
print("\nStudents scoring >85:\n", above_85)
```
